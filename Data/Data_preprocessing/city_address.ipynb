# Foodhub

## creating a new column

```python
data = pd.read_csv('foodhub.csv')
data.shape
```

```python
data.columns
```

```python
data['full_address'] = (
    data['Country'].fillna('') + ' ' +
    data['Region'].fillna('') + ' ' +
    data['City'].fillna('') + ' ' +
    data['Street'].fillna('') + ' ' +
    data['Number'].fillna('') + ' ' +
    data['Postcode'].fillna('')
)
```

## connection to database

```python
import mysql.connector

connection = mysql.connector.connect(
    host='p3nlmysql165plsk.secureserver.net',
    user='gholi_pour__',
    password='2024data_Wee%159159gholipour%$%',
    database='ph15928026353_'
)

cursor = connection.cursor()
```

```python
select_query = "SELECT id, Country, Region, City, Street, Number, Postcode FROM Foodhub_l1"
cursor.execute(select_query)
rows = cursor.fetchall()

```

```python
update_query = """
UPDATE Foodhub_l1
SET full_address = %s
WHERE id = %s
"""

for row in rows:
    id, Country, Region, City, Street, Number, Postcode = row
    full_address = (
        (Country or '') + ' ' +
        (Region or '') + ' ' +
        (City or '') + ' ' +
        (Street or '') + ' ' +
        (Number or '') + ' ' +
        (Postcode or '')
    ).strip()

    cursor.execute(update_query, (full_address, id))

    connection.commit()
    print(f"Updated row ID: {id} with full_address: {full_address}")

cursor.close()
connection.close()
```
