# lat long API

```python


def get_data(address, api_key="669f89cd51587199182297ejy53a146"):
    url = f"https://geocode.maps.co/search?q={address}&api_key={api_key}"
    session = requests.Session()
    retries = Retry(total=5, backoff_factor=1, status_forcelist=[502, 503, 504])
    session.mount('https://', HTTPAdapter(max_retries=retries))
    response = session.get(url)
    
    if response.status_code == 200:
        data = response.json()
        
        if data:
            lat = data[0].get("lat", None)
            lon = data[0].get("lon", None)
            return lat, lon
        else:
            return None, None
    else:
        print(f"Failed to retrieve data. HTTP Status code: {response.status_code}")
        return None, None
```

```python
import numpy as np


def update_csv_with_coordinates(file_path):
    df = pd.read_csv(file_path)
    df['Latitude'] = np.nan
    df['Longitude'] = np.nan
    # rows_to_process = df.iloc[0:15000]
    for index, row in df.iterrows():
        if pd.isna(row['Latitude']) or pd.isna(row['Longitude']):
            address = row['address']
            lat, lon = get_data(address)
            print(f"Processing address: {address}")

            if lat is not None and lon is not None:
                df.at[index, 'Latitude'] = lat
                df.at[index, 'Longitude'] = lon
                print(f"Updated: {address} -> Latitude: {lat}, Longitude: {lon}")
            else:
                print(f"Could not find coordinates for: {address}")
        else:
            print(f"Skipping address (already has coordinates): {row['Address']}")

    df.to_csv(file_path, index=False)
    print("CSV file has been updated.")

```
